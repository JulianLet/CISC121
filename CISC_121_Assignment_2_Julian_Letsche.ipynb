#  Problem 1 (Sum of Two Integers)




##1. Understanding

Adding two numbers without using + or - makes me think about binary addition. Because in binary there is only 0 and 1s we can just add by using the & operator. So we have to get the binary representation of both numbers and loop through each bit to add. Also a temp number to carry over.

The comparison of all three problems follows in problem 3.

##2. Peer Review & Draft Code

I looked at this: https://leetcode.com/problems/sum-of-two-integers/solutions/5375874/easy-solution-w-detailed-explanation-dry-run-c-java-beats-100/

```python
class Solution:
    def getSum(self, a: int, b: int) -> int:
        while b > 0:
            carry = a & b
            a = a ^ b
            b = carry << 1
        return a
```

## 3. Final Correct Code


```python
# Final working solution
# [AI Disclaimer example: Used ChatGPT with prompt “Does this fit for numbers -1000 till 1000” to see at edge cases]

class Solution:
    def getSum(self, a: int, b: int) -> int:

        MASK = 0xFFFFFFFF      #to restrict to 32-bits

        while b > 0:
            carry = (a & b) & MASK
            a = (a ^ b) & MASK
            b = (carry << 1) & MASK

        # if a is negative in 32-bit representation
        return a if a <= MASK else ~(a ^ MASK)
```

# Problems 4 (Reverse String)
--------

##1. Understanding

In a recursive way it would switch the outermost characters as a recursive function and send the inner into the function again.

The base case is if the substring is only 2 or 1 char long then we just return it.

There is a build in function to reverse strings which could be used instead.

##2. Peer Review & Draft Code

I looked at this: https://leetcode.com/problems/reverse-string/solutions/5244416/recursive-approach-in-python/

```python
class Solution:
    def reverseString(self, s: List[str]) -> None:
        def swap(left, right):
            if left >= right:
                return

            temp = s[left]
            s[left] = s[right]
            s[right] = temp
            swap(left+1, right-1)

        swap(0, len(s)-1)
```

## 3. Final Correct Code

```python
# Final working solution

class Solution:
    def reverseString(self, s: List[str]) -> None:
        def swap(left, right):
            if left >= right:
                return

            s[left], s[right] = s[right], s[left]
            swap(left+1, right-1)

        swap(0, len(s)-1)
```

# Problems 5 (Reverse Pairs)

----

##1. Understanding

so a pair is if arr[i] > arr[j] * 2 and i < j within the bounds of the array

so base case is if i == arr.length then we are finished

the recursive case loops from j = i+1 till the end of the array and count how often it is true then return that number + the same for i+1

this could just be done in a loop. instead of increasing i through recursion just increase it in an outer loop

##2. Peer Review & Draft Code

I looked at this: https://leetcode.com/problems/reverse-pairs/solutions/6188617/python-easy-to-understand-solution/

```python
class Solution:
    def reversePairs(self, nums: List[int]) -> int:
        def countPairs(i):
            if i == len(nums):
                return 0

            count = 0
            for j in range(i+1, len(nums)-1):
                if nums[i] > 2* nums[j]:
                    count += 1

            return count + countPairs(i+1)

        return countPairs(0)
```

## 3. Final Correct Code

```python
# Final working solution

class Solution:
    def reversePairs(self, nums: List[int]) -> int:
        def countPairs(i):
            if i == len(nums):
                return 0

            count = 0
            for j in range(i+1, len(nums)):
                if nums[i] > 2* nums[j]:
                    count += 1

            return count + countPairs(i+1)

        return countPairs(0)
```
