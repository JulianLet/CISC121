```python
class Solution:
    def reverse(self, x: int) -> int:
        negative = x < 0
        string_x = str(abs(x))

        reversed_string = string_x[::-1]
        reversed_int = int(reversed_string)

        if negative:
            reversed_int = -reversed_int

        if reversed_int > 2**31 - 1 or reversed_int < -2**31:
            return 0

        return reversed_int
```

```python
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        hash_tbl = {}

        i = 0
        for digit in nums:
            if target - digit in hash_tbl:
                return [i, hash_tbl[target - digit]] #if found return it
            hash_tbl[digit] = i #add it to hash if not found
            i = i + 1
```

I learned about the isalnum() function in python which makes it unnecessary to remove spaces and non alphanumeric numbers. The second solution posted is simmilar to my version from the int palindrom exercise earlier.



```python
class Solution:
    def isPalindrome(self, s: str) -> bool:
        i = 0
        j = len(s) - 1 #when in array consider 0 index

        while i < j:
            while not s[i].isalnum():
                i += 1
            while not s[j].isalnum():
                j -= 1

            if s[i] is not s[j]:
                return False

            i += 1
            j -= 1

        return True
```
