so for each char in the input string check the following order
1. if an open bracket push to stack

if 2 or 3 fail the conditions are not met so return false. else true

## 1. Understanding & Breakdown

there is given an int into the function and we have to test if it is a power of 3. A recursive way is:

- base case: if number is 0 it is divisible by three return true.
- else: check if number is divisible by 3. if no return false
- recursive: divide number by 3 and call function again.

## 2. Peer Review & Draft Code

I reviwed this solution: https://leetcode.com/problems/power-of-three/solutions/3242355/326-solution-with-step-by-step-explanati-oecx/?envType=problem-list-v2&envId=recursion

```python
class Solution:
    def isPowerOfThree(self, n: int) -> bool:
        #base case
        if (n == 0):
            return True

        if (n%3 != 0):
            return False

        return isPowerOfThree(n/3)
```

## 1. Understanding & Breakdown

## 2. Peer Review & Draft Code

I reviwed this solution: https://leetcode.com/problems/sort-colors/solutions/1078865/one-pass-on-time-o1-space-3-pointer-appr-ja2i/

this is something like i had in mind with three pointers. I learned that you can swap without temp in python
        

```python
class Solution:
    def sortColors(self, nums: List[int]) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        left = 0
        middle = 0
        right = len(nums) - 1

        while middle <= right:
            if nums[middle] < nums[left]:
                nums[left], nums[middle] = nums[middle], nums[left]
                left += 1
            elif nums[middle] > nums[right]:
                nums[right], nums[middle] = nums[middle], nums[right]
                right -= 1
            else:
                middle += 1

        return nums

```
